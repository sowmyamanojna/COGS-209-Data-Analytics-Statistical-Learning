```python
# https://pandas.pydata.org/pandas-docs/dev/reference/api/pandas.NamedAgg.html
num_data_points = pd.NamedAgg(column="activity", aggfunc="count")
avg_act = pd.NamedAgg(column="activity", aggfunc="mean")
max_act = pd.NamedAgg(column="activity", aggfunc="max")

# Loop through all conditions
for condn_fname in condn_files:
    # Load a condition file
    df_act_condn = pd.read_csv(CDPATH+condn_fname)
    df_act_condn.head(10)
    df_new = df_act_condn.groupby("date").agg(num_data_points=num_data_points, avg_act=avg_act, max_act=max_act)
    display(df_new)
```

```python
# Same as above: take only days with all activity and threshold
threshold = 50
num_samples = 0

# Loop through all conditions
for condn_idx in range(len(condn_files)):
    condn_fname = "condition_"+str(condn_idx+1)+".csv"
    # Load a condition file
    df_act_condn = pd.read_csv(CDPATH+condn_fname)
    df_act_condn.head(10)
    df_new = df_act_condn.groupby("date").agg(num_data_points=num_data_points, avg_act=avg_act, max_act=max_act)
    
    # Masking
    cleaned_df_new = df_new[df_new["num_data_points"] == 1440]
    cleaned_df_new = cleaned_df_new[cleaned_df_new["avg_act"] > threshold]
    # display(cleaned_df_new)
    print("Condition:", condn_fname.split("_")[-1][:-4])
    print(cleaned_df_new.shape)
    num_samples += cleaned_df_new.shape[0]
```

```python
num_samples
```

```python
# Stack the dataset to create a matrix of size num_samples*1440 for the control group
threshold = 50

for condn_idx in trange(len(condn_files)):
    condn_fname = "condition_"+str(condn_idx+1)+".csv"
    # Load a condition file
    df_act_condn = pd.read_csv(CDPATH+condn_fname)
    df_act_condn.head(10)
    df_new = df_act_condn.groupby("date", as_index=False).agg(num_data_points=num_data_points, avg_act=avg_act, max_act=max_act)
    
    # Masking
    cleaned_df_new = df_new[df_new["num_data_points"] == 1440]
    cleaned_df_new = cleaned_df_new[cleaned_df_new["avg_act"] > threshold]
    # display(cleaned_df_new)t
    
    # get all the final dates
    for date_idx, date in enumerate(cleaned_df_new["date"]):
        tmp_data = df_act_condn[df_act_condn["date"] == date]
        tmp_data = tmp_data["activity"].to_numpy().reshape(1, 1440)
        if condn_idx == 0 and date_idx == 0:
            X_condn = tmp_data
        else:
            X_condn = np.vstack([X_condn, tmp_data])
        
X_condn.shape
```

```python
# Doing the same thing on control datapoints
# Stack the dataset to create a matrix of size num_samples*1440 for the control group
threshold = 50

cntlr_files = os.listdir(CNPATH)

for cntlr_idx in trange(len(cntlr_files)):
    cntlr_fname = "control_"+str(cntlr_idx+1)+".csv"
    # Load a condition file
    df_act_condn = pd.read_csv(CNPATH+cntlr_fname)
    df_act_condn.head(10)
    df_new = df_act_condn.groupby("date", as_index=False).agg(num_data_points=num_data_points, avg_act=avg_act, max_act=max_act)
    
    # Masking
    cleaned_df_new = df_new[df_new["num_data_points"] == 1440]
    cleaned_df_new = cleaned_df_new[cleaned_df_new["avg_act"] > threshold]
    # display(cleaned_df_new)t
    
    # get all the final dates
    for date_idx, date in enumerate(cleaned_df_new["date"]):
        tmp_data = df_act_condn[df_act_condn["date"] == date]
        tmp_data = tmp_data["activity"].to_numpy().reshape(1, 1440)
        if cntlr_idx == 0 and date_idx == 0:
            X_cntlr = tmp_data
        else:
            X_cntlr = np.vstack([X_cntlr, tmp_data])
        
X_cntlr.shape
```
