```python
import os
import json
import pycurl
import numpy as np
import pandas as pd
from tqdm.notebook import trange, tqdm

try:
    os.mkdir("songs")
except:
    pass
```

```python
for year in range(2017, 2022):
    df = pd.read_csv("csv/"+str(year)+".csv", index_col=0)
    df["Ranking"] = list(range(1, df.shape[0]+1))
    df = df[["Ranking", "Song Names", "Artists", "Song IDs", "Streams"]]
    
    display(df.head())
    
    data_list = []
    
    for i in trange(df.shape[0]):
        c = pycurl.Curl()
        c.setopt(pycurl.URL, "https://api.spotify.com/v1/audio-features/"+df["Song IDs"][i])
        c.setopt(pycurl.HTTPHEADER, ["Authorization: Bearer  BQBBay6Ku8a8123q2wPsc9WDf9ZLkjsyAEBChaiE1esv00dTHE5tog1MpH-ePBZz4NfmxKmAdzxmvFgXNMLm0KcWHJUbmTfUQsUhjZqo6qKpiPRT_Y1B"])
        with open("songs/"+df["Song Names"][i]+'.json', 'wb') as f:
            c.setopt(c.WRITEFUNCTION, f.write)
            c.perform()
            
        f = open("songs/"+df["Song Names"][i]+".json")
        data = json.load(f)
        f.close()

        data_list.append(data)
        
    df_song = pd.DataFrame(data_list, index=range(df.shape[0]))
    df_song.to_csv("csv/"+str(year)+"song_features.csv")
    
    df = pd.concat([df, df_song], axis=1)
    df.to_csv("csv/"+str(year)+"_final.csv")
    display(df.head())
```

```python
# Additional fine tuning, cleaning the data
for year in range(2017, 2022):
    df = pd.read_csv("csv/"+str(year)+"_final.csv", index_col=0)
    column_names = df.columns
    column_names = [i.title() for i in column_names]
    df.rename(columns={i:j for i,j in zip(df.columns, column_names)}, inplace=True)
    df.rename(columns={"Time_Signature":"Time Signature"}, inplace=True)
    df.drop(["Type", "Id", "Uri", "Analysis_Url", "Track_Href", "Duration_Ms"], axis=1, inplace=True)
    df.to_csv("csv/"+str(year)+"_complete.csv")
    display(df.head())
```
